## Filtros

### Para filtrar as respostas dadas por mulheres

```python
answers[answers["Gender"] == "Woman"]
```

### Para aceder ao nível de educação das mulheres

```python
answers.loc[answers["Gender"] == "Woman", "EdLevel"]
```

### Para aceder ao nível de educação das mulheres cuja idade está entre os 18 e os 24 anos

```python
answers.loc[(answers["Gender"] == "Woman") & (answers["Age"] == "18-24 years old"), "EdLevel"]
```

### Para aceder ao nível de educação das pessoas que são mulheres ou cuja idade está entre os 18 e os 24 anos

```python
answers.loc[(answers["Gender"] == "Woman") | (answers["Age"] == "18-24 years old"), "EdLevel"]
```

### Para aceder ao nível de educação das pessoas que não são mulheres e a cuja idade não está entre os 18 e os 24 anos

```python
answers.loc[~((answers["Gender"] == "Woman") | (answers["Age"] == "18-24 years old")), "EdLevel"]
```

### Para aceder ao país, às linguagens com que já trabalharam e à compensação total das pessoas cuja compensação total é superior a 70000$

```python
answers.loc[answers["CompTotal"] > 70000, ["Country", "LanguageHaveWorkedWith", "CompTotal"]]
```

### Para aceder às respostas cujo país toma o valor de Alemanha, Suiça ou Itália

```python
countries = ["Germany", "Switzerland", "Italy"]

answers.loc[answers["Country"].isin(countries)]
```

### Para aceder às respostas cujo campo com as linguagens com que já trabalharam inclui Python

```python
answers.loc[answers["LanguageHaveWorkedWith"].str.contains("Python", na = False), "LanguageHaveWorkedWith"]
```

## Limpar dados

### Para criar um novo DataFrame com dados em falta

```python
people = {
    "first": ["Ana", "Bruno", "Hugo", "Pedro", np.nan, None, "NA"],
    "last": ["Fernandes", "Ramalho", "Pimentel", "Santos", np.nan, np.nan, "Missing"],
    "email": ["AnaFernandes@gmail.com", "BrunoRamalho@gmail.com", "HugoPimentel@gmail.com", None, np.nan, "Anonimo@gmail.com", "NA"],
    "age": [34, 67, 45, 45, None, None, "Missing"]
}
```

```python
people_df = pd.DataFrame(people)
```

```python
people_df
```

### Para eliminar as linhas que tenham um ou mais campos com _NaN_ ou _None_

```python
people_df.dropna()
```

### Para eliminar as linhas que tenham todos os campos com _NaN_ ou _None_

```python
people_df.dropna(how = "all")
```

### Para eliminar as colunas que tenham _NaN_ ou _None_

```python
people_df.dropna(axis = "columns")
```

### Para eliminar as linhas cujo email seja _NaN_ ou _None_

```python
people_df.dropna(subset=["email"])
```

### Para eliminar as linhas cujo último nome ou email sejam _NaN_ ou _None_

```python
people_df.dropna(subset=["last", "email"])
```

### Para substituir _NA_ por _NaN_

```python
people_df.replace("NA", np.nan, inplace=True)
```

```python
people_df
```

### Para substituir _Missing_ por _NaN_
